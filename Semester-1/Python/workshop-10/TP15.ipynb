```python
def ActeursDebutants(c):
  c.execute("SELECT a.nom, a.prenom, avg(f.salaire) as moyenne_salaire FROM acteur a join filmographie f on f.idActeur=a.idActeur WHERE idActeur IN (SELECT idActeur FROM filmographie) group by a.idActeur ")
  for ligne in c.fetchall():
    print(ligne)
```

Q14

```python
def ActeursMemeSalaire(c):
  c.execute("SELECT distinct a1.nom,a1.prenom, a2.nom, a2.prenom FROM acteur a1, acteur a2 WHERE a1.idActeur <> a2.idActeur AND a1.salaire = a2.salaire")
  for ligne in c.fetchall():
    print(ligne)
```

```python
def SalaireDollarToDirham(c):
  c.execute("SELECT a.nom, a.prenom, f.salaire*10 as salaire_dirham FROM acteur a join filmographie f on f.idActeur=a.idActeur")
  for ligne in c.fetchall():
    print(ligne)
```
