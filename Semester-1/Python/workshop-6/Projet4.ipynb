étape 1

```python
import sympy
import numpy as np
```

```python
fonction_str = input("Entrez votre fonction f(x,y) : ")
```

```python
x, y = sympy.symbols('x y')
```

```python
expression_sympy = sympy.sympify(fonction_str)
```

```python
print(expression_sympy)
```

étape 2

```python
f_numpy = sympy.lambdify([x, y], expression_sympy, 'numpy')
```

```python
print(type(f_numpy))
```

```python
x, y = 2, 3
result = f_numpy(x, y)
print(result)
```

étape 3

```python
#DEF du domaine
x_min_domain = -10
x_max_domain = 10
y_min_domain = -10
y_max_domain = 10
nb_points = 100
```

```python
#vect linéaire
x_vals = np.linspace(x_min_domain, x_max_domain, nb_points)
y_vals = np.linspace(y_min_domain, y_max_domain, nb_points)
```

```python
#matrice 2D
X_grid, Y_grid = np.meshgrid(x_vals, y_vals)
```

étape 4

```python
# application des fcs
Z_results = f_numpy(X_grid, Y_grid)
```

```python
# la valeur min
min_val = np.min(Z_results)
print(f"valeur minimale : {min_val:.4f}")
```

```python
# les indices du min
indices = np.unravel_index(np.argmin(Z_results), Z_results.shape)
print(f"les indices du min : {indices}")
```

```python
# les coordonnées
x_min = x_vals[indices[1]]  # cest 1 car indince de collone = pos de x
y_min = y_vals[indices[0]]

print(f"Position du minimum : x = {x_min:.5f}, y = {y_min:.5f}")
```

étape 5

```python
print(f"la Fonction : {fonction_str}")
print(f"le min est {min_val:.6f} et les coordonées sont (x, y) = ({x_min:.4f}, {y_min:.4f})")
```
