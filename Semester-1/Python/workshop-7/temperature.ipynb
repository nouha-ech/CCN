Q3 création nv dataframe

```python
df['month'].dtype
```

```python
dff = df[( (df["month"] == 3) | (df["month"] == 6) | (df["month"] == 9) | (df["month"] == 12) ) & (df["Country"] != 'EAST')]
```

```python
dff
```

Q4 calcul numpy

```python
import numpy as np
```

```python
dfi = dff.pivot_table(
        index='City',
        columns='month',
        values='AverageTemperatureFahr'
    )
```

```python
dfi
```

```python
data= dfi.to_numpy()
```

```python
avg = data.mean(axis=0)
```
